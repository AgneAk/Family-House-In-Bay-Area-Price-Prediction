Calculating price predictions on data using XGBoostRegressor

```python
import pandas as pd

import numpy as np
```

```python
df_for_predictions = pd.read_csv('/Users/baytech11/Desktop/Agnes/Springboard/Capstone Project/prepared data/final_data_unsclaed')
```

```python
df_for_predictions.head(3)
```

```python
import xgboost as xg
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
X= df_for_predictions.drop("Price per sqf ($)", axis=1).values

y = df_for_predictions['Price per sqf ($)'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 50)
```

```python
xgb = XGBRegressor()
```

```python
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```

```python
xgb_predict = xgb.predict(X_test[0:10])
```

```python
xgb_predict
```

```python
predictions = pd.DataFrame(X_test[0:10], columns = [df_for_predictions.drop("Price per sqf ($)", axis=1)])
```
